```python
# Adding style to yhe graph
from matplotlib import pyplot as plt
from matplotlib import style

style.use('bmh')

x = [5,8,10]
y = [12,16,6]

x1 = [6,15,7]
y1 = [6,9,11]

plt.plot(x,y,'r',label="line one",linewidth=5)
plt.plot(x1,y1,'g',label="line two",linewidth=5)

plt.title("Epic line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()

plt.grid(True,color="k",linewidth=0.2)
plt.show()
```
